```python
def str_to_list(x):
    return ast.literal_eval(x)
```

```python
ted['rating_list'] = ted['ratings'].apply(str_to_list)
```

```python
ted.loc[0, 'rating_list']
```

```python
ted.loc[0, 'rating_list'][0]['count']
```

```python
def count_sum(x):
    num = 0
    length = len(x)
    for i in range(length):
        num = num + x[i]['count']
    return num
```

```python
ted['rating_counts'] = ted['rating_list'].apply(count_sum)
```

```python
ted.head()
```

```python
import datetime
```

```python
def time_diff(x):
    newdf = datetime.datetime.now()- x
    return newdf.days
```

```python
ted['timediff'] = ted['re_published_date'].apply(time_diff)
```

```python
ted['avg_ratings'] = ted['rating_counts']/ted['timediff']
```

```python
ted.head()
```

```python
def funny_counts(x):
    num = 0
    length = len(x)
    for i in range(length):
        if x[i]['name'] == 'Funny':
            num = num + x[i]['count']
            
    return num
```

```python
funny_counts(ted.loc[0, 'rating_list'])
```

```python
ted['funny_counts'] = ted['rating_list'].apply(funny_counts)
```
